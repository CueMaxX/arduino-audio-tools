## Effects

```c++
#pragma cling add_include_path("/Users/pschatzmann/Dev/Arduino/libraries/arduino-audio-tools/src")
#include "AudioTools.h"
#include "AudioLibs/Jupyter.h"
```

We set up some effects based on a sine generator

```c++
SineWaveGenerator<int16_t> sine;
AudioEffects<SineWaveGenerator<int16_t>> effects(sine);
GeneratedSoundStream<int16_t> in(effects); 
auto cfg = in.defaultConfig();
cfg.channels = 1;
cfg.sample_rate = 44100;
in.begin(cfg);
sine.begin(cfg, 261);
```

```c++
JupyterAudio audio("sine.wav", in, 600, 1024);
audio
```

```c++
effects.clear();
Boost effect(0.1);
effects.addEffect(effect);

JupyterAudio audio("sine-boost.wav", in, 600, 1024);
audio
```

```c++
effects.clear();
Distortion effect(10000, 32767);
effects.addEffect(effect);

JupyterAudio audio("sine-distortion.wav", in, 600, 1024);
audio
```

```c++
effects.clear();
Fuzz effect(4.0, 10000);
effects.addEffect(effect);

JupyterAudio audio("sine-fuzz.wav", in, 600, 1024);
audio
```

```c++
effects.clear();
Tremolo effect(400, 50, cfg.sample_rate);
effects.addEffect(effect);

JupyterAudio audio("sine-tremolo.wav", in, 600, 1024);
audio
```

```c++
effects.clear();
Delay effect(200, 0.5, 0.5, cfg.sample_rate);
effects.addEffect(effect);

JupyterAudio audio("sine-delay.wav", in, 600, 1024);
audio
```

```c++
effects.clear();
PitchShift effect(1.8);
effects.addEffect(effect);

JupyterAudio audio("sine-pitchshift.wav", in, 600, 1024);
audio
```
