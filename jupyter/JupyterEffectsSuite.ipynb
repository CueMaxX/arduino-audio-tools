## Effects

```c++
#pragma cling add_include_path("/Users/pschatzmann/Dev/Arduino/libraries/arduino-audio-tools/src")
#include "AudioTools.h"
#include "AudioLibs/Jupyter.h"
#include "AudioLibs/AudioEffectsSuite.h"
```

We set up some effects based on a sine generator

```c++
ModulationBaseClass modulation;
modulation.setSine();
SoundGeneratorModulation<int16_t> gen(modulation);
AudioEffects<SineWaveGenerator<int16_t>> effects(gen);
GeneratedSoundStream<int16_t> in(effects); 
auto cfg = in.defaultConfig();
cfg.channels = 1;
cfg.sample_rate = 44100;
in.begin(cfg);
sine.begin(cfg, 261);
```

```c++
JupyterAudio audio("sine.wav", in, 600, 1024);
audio
```

```c++
effects.clear();
SimpleDelay effect(50);
effects.addEffect(effect);

JupyterAudio audio("sine-simpledelay.wav", in, 200, 1024);
audio
```

```c++
effects.clear();
SimpleChorus effect(cfg.sample_rate);
effect.setupChorus(cfg.sample_rate);
effect.setSwing(0.5);
effect.setBase(1.0);
effects.addEffect(effect);

JupyterAudio audio("sine-chorus.wav", in, 600, 1024);
audio
```

```c++
effects.clear();
FilteredDelay effect(1000);
effect.setDelayGain(0.6);
effect.setFeedbackGain(0.5);
effects.addEffect(effect);

JupyterAudio audio("sine-filtereddelay.wav", in, 600, 1024);
audio
```

```c++
effects.clear();
SimpleFlanger effect(cfg.sample_rate);
effect.setupSimpleFlanger(cfg.sample_rate);

JupyterAudio audio("sine-simpleflanger.wav", in, 600, 1024);
audio
```
